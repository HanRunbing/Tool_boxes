# Merge

## concat

```python
df_mer = pd.read_csv("/content/drive/MyDrive/coding_practise/for_merge.csv")
df_mer.head()
```

```python
# Concatenate pandas objects along a particular axis with optional set logic along the other axes. more often to use add new row the end or col in the right
A = df[['gene','x','y']]
B = df_mer[['cluster','density']]
pd.concat([A,B],axis = 1).head()  # add additional info in the right col
```

```python
pd.concat([A,B],keys = ['segmentation','cell status'],axis = 1).head()  # keys is to mark the belonings
```

```python
Blank1_50 = df_2.loc['Blank1',:][1:50]
Blank10_50 = df_2.loc['Blank10',:][1:50]
```

```python
pd.concat([Blank1_50,Blank10_50], keys = ['Blank1_50','Blank10_50'])
```
